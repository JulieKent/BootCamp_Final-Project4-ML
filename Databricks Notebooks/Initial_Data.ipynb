```
%sql
create or replace view cx.interactions as
SELECT cd.conversationStart,
    ca.conversationId,
    ca.mediaType,
    ca.wrapUpText,
    COUNT(DISTINCT ca.conversationId) as offered
FROM reporting.genesys__conversations_aggregates_wide AS ca
JOIN edw.genesys__conversations_details as cd ON ca.conversationId = cd.conversationId
JOIN
  (SELECT try_subtract(max(date(conversationStart)),interval '12' month) AS minDate, CURRENT_DATE as maxDate FROM edw.genesys__conversations_details) AS date_range
ON cd.conversationStart between date_range.minDate and date_range.maxDate
WHERE ca.originatingDirection = 'inbound'
AND ca.wrapUpText is not null
GROUP BY cd.conversationStart, ca.conversationId, ca.mediaType, ca.wrapUpText;
```

```
%sql
SELECT * FROM cx.interactions;
```
