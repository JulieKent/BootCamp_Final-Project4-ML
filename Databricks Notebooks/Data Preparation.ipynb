```
%sql
-- Sum the count of unique conversations grouped by date and media type and joined by the date to the difot score
Create or replace view int_by_MediaType as
SELECT
  DATE(i.conversationStart) AS intDate,
  i.mediaType,
  COUNT(DISTINCT i.conversationID) AS offerCount,
  d.Score 
FROM interactions as i
JOIN difot_scores AS d ON DATE(i.conversationStart) = d.Date
WHERE DATE(i.conversationStart) >= (select min(Date) from difot_scores) -- filter interactions by the earliest date in the difot_scores table
GROUP BY intDate, i.mediaType, d.Score
ORDER BY intDate ASC;

SELECT * FROM int_by_MediaType;
```

```
%sql
-- Sum the count of unique conversations grouped date and join by the date to the difot score
Create or replace Table daily_interactions_difot as
SELECT
  DATE(i.conversationStart) AS intDate,
  COUNT(DISTINCT i.conversationID) AS offerCount,
  d.Score 
FROM interactions as i
JOIN difot_scores AS d ON DATE(i.conversationStart) = d.Date
WHERE DATE(i.conversationStart) >= (select min(Date) from difot_scores) -- filter interactions by the earliest date in the difot_scores table
GROUP BY intDate, d.Score
ORDER BY intDate ASC;

SELECT * FROM daily_interactions_difot;
```
